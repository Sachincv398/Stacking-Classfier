```python
import pandas as pd
import numpy as np
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df= pd.read_csv("Bank_churn_modelling.csv")
```

```python
df.head()
```

```python
df = df.drop('RowNumber', axis = 1)
```

```python
df = df.drop(['Surname','CustomerId'],axis = 1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
```

```python
df['Geography']=label.fit_transform(df['Geography'] )
```

```python
df
```

```python
df = pd.get_dummies(df,columns = ['Gender'], prefix = 'R')
```

```python
df
```

```python
df
```

```python
algo = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= df.drop('Exited', axis = 1)
y = df['Exited']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8)
```

```python
base_learner = [('clf1',KNeighborsClassifier(n_neighbors=5)),('clf2',DecisionTreeClassifier(criterion ='entropy'))]
```

```python
stacking = StackingClassifier(estimators = base_learner, final_estimator= algo)
```

```python
stacking.fit(xtrain, ytrain)
```

```python
y_pred = stacking.predict(xtest)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest, y_pred))
```
